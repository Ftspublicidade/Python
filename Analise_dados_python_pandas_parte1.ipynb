### Agora vou mostrar o método loc que é utilizado para visualizar informações do dataset, este método recebe uma lista por parâmetro e retorna o resultado da consulta. Abaixo todas as informações referentes ao continente oceania:

```python
oceania = df.loc[df["continent"] == "Oceania"]
oceania.head()
```

### E se quisermos agrupar os dados por uma determinada coluna? o código abaixo retorna o agrupamento por continente e retorna a quantidade de países por continente utilizando o método nunique:

```python
df.groupby("continent")["country"].nunique()
```

### Para cada ano em nossos dados, qual era a expectativa de vida média? Segue código abaixo:

```python
df.groupby("year")["lifeExp"].mean()
```
